````
correlation <- cor(industrial, tax)

if(correlation > 0.5 | correlation < -0.5) {
  cat('Korelasi Tax Dan Industrial Adalah', correlation, 'Sehingga Merupakan Korelasi Kuat')
} else {
  cat('Korelasi Tax Dan Industrial Adalah', correlation, 'Sehingga Merupakan Korelasi Lemah')
}
````
